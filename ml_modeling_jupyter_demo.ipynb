# STEP-2: 数据预处理 Data pre-processing (N/A replacement)

```python
df = df.drop(["phone number"],axis = 1)
```

```python
df=df.fillna(df.mean())
df
```

# STEP-3. 类别型特征One-hot encoding

```python
df.columns
```

```python
#one hot encoding.
l = ['state',"area code","international plan","voice mail plan"]
def one_hot_encoding(df, l):
	for n in l:
		df_ohe = pd.get_dummies(df[n], prefix=n)
		print(n, df_ohe.columns)
		df = df.drop([n],axis=1)
		df = pd.concat([df, df_ohe], axis = 1)
	return df
df = one_hot_encoding(df,l)
```

```python
df.columns
df
```

# STEP-4 训练测试数据集分割 Train & test Split

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df,random_state=12,test_size=0.2)
```

```python
train_X_df, train_y = train_df.drop(["churn"],axis=1), train_df["churn"].to_numpy()
test_X_df, test_y = test_df.drop(["churn"],axis=1), test_df["churn"].to_numpy()
```

# STEP-5 数据normalization

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
train_X_df_scaled = ss.fit_transform(train_X_df)
test_X_df_scaled = ss.transform(test_X_df)
```

# STEP-6 训练模型Training model

```python
# model training and prediction
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(train_X_df_scaled, train_y)
pred_y_LR = clf.predict(test_X_df_scaled)
pred_y_prob_LR = clf.predict_proba(test_X_df_scaled)[:,1]
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(train_X_df_scaled, train_y)
pred_y_DT = clf.predict(test_X_df_scaled)
pred_y_prob_DT = clf.predict_proba(test_X_df_scaled)[:,1]
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(train_X_df_scaled, train_y)
pred_y_GBDT = clf.predict(test_X_df_scaled)
pred_y_prob_GBDT = clf.predict_proba(test_X_df_scaled)[:,1]
```

# STEP-7. 模型评估Evaluation

```python
from sklearn import metrics
from sklearn.metrics import precision_score,recall_score
def model_evluation(test_y, pred_y, pred_y_prob):
	fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_y_prob, pos_label=1)
	print("AUC = " + str(metrics.auc(fpr, tpr)))
	print("Precision = " + str(precision_score(test_y, pred_y, average='macro')))
	print("Recall = " + str(recall_score(test_y, pred_y, average='macro')))
```

```python
model_evluation(test_y, pred_y_LR, pred_y_prob_LR)
```

```python
model_evluation(test_y, pred_y_DT, pred_y_prob_DT)
```

```python
model_evluation(test_y, pred_y_GBDT, pred_y_prob_GBDT)
```
